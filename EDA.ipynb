```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
happy = pd.read_csv("./happy.csv", na_values = 'Not applicable', low_memory=False) #Importing the file with "Non applicable data as NA"
happy2=happy.dropna()
list(happy.columns.values) #Displays the variable names
happy2.shape #Displays the dimensions of a data frame
```

General Social Survey (GSS NORC) was collecting data since 1972 from thousands of people all across the U.S. For the purposes of this analysis, we used data only from 2016 since this year has the least number of missing data. Overall, 2867 observations of 12 variables were collected, among which only one variable ("education") is continious . Each observation represents the responses of one person.

3. Let's compare the frequencies of the categorical data and transform variables "age" and "children" to the continious variables.

```python
happiness_bar=happy2['happiness'].value_counts().plot(kind='bar')
happiness_bar.set_title("Happiness", fontsize=18)
```

```python
ms=happy2['marriage_satisf'].value_counts().plot(kind='bar')
ms.set_title('Satisfaction in marriage', fontsize=18)
```

```python
sexbar=happy2['sex'].value_counts()
labels = 'Female', 'Male'
sizes = [53, 47]
explode = (0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
labor_status_bar=happy2['labor_status'].value_counts().plot(kind='bar')
labor_status_bar.set_title("Participants' labor status", fontsize=18)
```
